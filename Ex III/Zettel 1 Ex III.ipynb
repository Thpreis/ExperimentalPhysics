Schwarzkörper Spektrum in Abhängigkeit der Wellenlänge und der Frequenz geplottet


```python
%matplotlib inline
##Plot u(f)
#Konstanten
h = 6.626e-34
k = 1.38e-23
c = 299792458.
def Spektrum_Schwarzer_Körper_frequency(f, T, h):
    return (8*np.pi*h*f**3)/((c**3)*(np.exp(h*f/(k*T))-1))
f = np.linspace(100, 1000000000000000, 100000)
h = 6.626e-34
plt.plot(f, Spektrum_Schwarzer_Körper_frequency(f, 5000, h), linewidth=2)
plt.plot(f, Spektrum_Schwarzer_Körper_frequency(f, 4500, h), linewidth=2)
plt.plot(f, Spektrum_Schwarzer_Körper_frequency(f, 4000, h), linewidth=2)
plt.plot(f, Spektrum_Schwarzer_Körper_frequency(f, 800, 6.626e-35), linewidth=2)


plt.ylim(1e-18, 1.05e-15)
plt.xlabel("Frequenz[Hz]")
plt.ylabel("Strahlungsdichte[W m^-2]")
plt.legend(["T = 5000K","T = 4500K","T = 4000K","T = 800K \nh = 6*10^-35 "])
plt.savefig("Spektrum_Frequenz.png")
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#Konstanten
h = 6.626e-34
k = 1.38e-23
c = 299792458.

def Spektrum_Schwarzer_Körper_wavelength(l, T, x):
    return (8*np.pi*x)/((l**5)*(np.exp(x/(l*k*T))-1))
x = 1.986e-25
x_new = 1.986e-26
l = np.linspace(5e-8, 3e-6, 1000)
plt.plot(l, Spektrum_Schwarzer_Körper_wavelength(l, 5000, x), linewidth=2)
plt.plot(l, Spektrum_Schwarzer_Körper_wavelength(l, 4500, x), linewidth=2)
plt.plot(l, Spektrum_Schwarzer_Körper_wavelength(l, 4000, x), linewidth=2)
plt.plot(l, Spektrum_Schwarzer_Körper_wavelength(l, 800, x_new), linewidth=2)


plt.ylim(0., 600000.)
plt.xlabel("Wellenlänge[m]")
plt.ylabel("Strahlungsdichte[W m^-2]")
plt.legend(["T = 5000K","T = 4500K","T = 4000K","T = 800K \nh = 6*10^-35 "])
plt.savefig("Spektrum_Wellenlänge.png")
```
