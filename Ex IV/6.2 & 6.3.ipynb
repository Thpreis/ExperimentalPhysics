```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
theta = np.linspace(0,45, 900)
def Rate(x):
    return 3.047e-5*(np.sin(x*np.pi/360))**(-4)
plt.plot(theta, Rate(theta), color="red", label="Reaktionsrate im Target")
plt.title("Reaktionsrate im Target abhängig vom Streuwinkel")
plt.yscale("log")
```
