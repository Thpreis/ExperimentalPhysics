```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
y = 10.
wo = 50.
def Spektrum(w):
    return (y**2 + wo - w)/((y**2+(wo-w)**2)**2)

w = np.linspace(0., 100., 100)

plt.plot(w, Spektrum(w), linewidth=3, color="red")
plt.title("Spektrum")
plt.xlabel("Frequenz w [Hz]")
plt.ylim(0., 0.011)
plt.ylabel("Spektrum q(w)")
plt.savefig("Spektrum.pdf")
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
y = 30.
wo = 50.

def Phase(w):
    x = -y/(wo-w)
    return np.arctan(x)

w = np.linspace(0., 100., 100)

plt.plot(w, Phase(w), linewidth=3, color="red")
plt.title("Phase")
plt.xlabel("Frequenz w [Hz]")
plt.savefig("Phase1.pdf")
```
