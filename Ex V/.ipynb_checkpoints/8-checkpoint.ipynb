```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#constants
k=1.38e-23
h=6e-34
theta=345
m=1.05e-25
a=3.61e-10

#Termperature interval
T = np.linspace(0.01,100,100000)
```

```python
def CVtot(T):
    return np.pi**2 *k**2 *m*a**2 *1/(18**(2/3)) *1/h**2 *T + 12*np.pi**4 *k*T**3 *1/5 *1/(theta**3)
def CVelec(T):
    return np.pi**2 *k**2 *m*a**2 *1/(18**(2/3)) *1/h**2 *T
def CVphon(T):
    return 12*np.pi**4 *k*T**3 *1/5 *1/(theta**3)
```

```python
plt.plot(T, CVtot(T), label="Gesamtwärmekapazität",lw=3)
plt.plot(T,CVelec(T), label="Elektronischer Beitrag", lw=1.5, color="red")
plt.plot(T, CVphon(T), label="Phononischer Beitrag",lw=1.5)
plt.title("Wärmekapazität von Kupfer")
plt.xlabel("ln(T) [K]")
plt.ylabel("$ln(C_V)$ $[J/K ]$")
plt.xscale("log")
plt.yscale("log")
plt.legend(loc="best")
plt.savefig("Wärmekapazität von Kupfer.pdf")
```
