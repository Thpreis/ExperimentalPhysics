```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#1 Ist amorphe Struktur, 2 ist kristalline Struktu
n01=80
n02=81
r_n = np.linspace(1,17,17)
#n_2 = np.array([1,1,1,5,5,6,6,10,18,18,18,26,34,34,42,46,58])
#n_1 = np.array([1,1,1,5,7,7,7,12,18,20,24,30,34,38,42,50,59])
n_2 = np.array([1,0,0,4,0,1,0,4,8,0,0,8,8,0,8,4,12])
n_1 = np.array([1,0,0,4,2,0,0,5,6,2,4,6,4,4,4,8,9])
def g(n,n0):
    return n/n0
```

```python
plt.plot(r_n, g(n_2,n02), label="Kristall", lw=2, color="red")
plt.plot(r_n, g(n_1,n01), label="Amorpher Festkörper",lw=1.5)
plt.title("Teilchenverteilung beider Festkörper")
plt.xlabel("Radius vom Aufatom $r=n\cdot R$, [R] ")
plt.ylabel("Radiale Paarverteilungsfunktion $g(r)$")
plt.legend(loc="upper left")
plt.savefig("Verteilungsfunktion1.pdf")

```

```python


from scipy.interpolate import interp1d
x=np.linspace(1,17,1000)

f = interp1d(r_n, n_2, kind="cubic")
g = interp1d(r_n, n_1, kind="cubic")
plt.plot(x,f(x), label="Kristall", lw=1.5)
plt.plot(x,g(x), label="Amorpher Festkörper", lw=1.5)
plt.title("Teilchenverteilung beider Festkörper")
plt.xlabel("Radius vom Aufatom $r=n\cdot R$, [R] ")
plt.ylabel("Radiale Paarverteilungsfunktion $g(r)$")
plt.legend(loc="upper left")
plt.savefig("Verteilungsfunktion.pdf")

```
