```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
a=3 #Angstrom
N=10
v=2000 #m/s
q = np.linspace(-np.pi/a,np.pi/a, 10000)
```

```python

def w(q):
    return 2/a*v*np.abs(np.sin(q*a/2))
```

```python
plt.plot(q,w(q))
```

```python
#### 7.3
b=4.08
def wakustisch(q):
    return np.sqrt(18.5*(1740/b)**2*(1-np.sqrt(1-4*0.21*(np.sin(q*b/2))**2)))
def woptisch(q):
    return np.sqrt(18.5*(1740/b)**2*(1+np.sqrt(1-4*0.21*(np.sin(q*b/2))**2)))
```

```python
plt.plot(q, wakustisch(q), label="akustisch")
plt.plot(q, woptisch(q), label="optisch")
plt.legend(loc="lower left")
plt.xlabel("q")
plt.ylabel("Dispersionsrelation w(q)")
```
