```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

def vph(k, m):
    return (np.sqrt(299792458**2 * k**2+( m**2 * 299792458**4 )/((6e-34)**2)))/k

def vg(k, m):
    return k*(299792458**2)/(np.sqrt(299792458**2 * k**2+( m**2 * 299792458**4 )/((6e-34)**2)))

k = np.linspace(1e-12, 1e6, 200)


plt.plot(k, vph(k, 1e-32), lw=2, color="navy")
plt.ylim(1e-12, 5e8)
plt.legend(["$Phasengeschwindigkeit$"+"  ""$ v_{ph}$"], loc="best")
plt.xlabel("$Wellenvektor$"+"  "+"$k [m]$")
plt.ylabel("$Geschwindigkeit$" +"  "+"$[m/s]$")
plt.title("Verlauf der Phasengeschwindigkeit")
#plt.savefig("Phasengeschw.pdf")
```

```python

%matplotlib inline
k = np.linspace(1., 5e23, 5000)
plt.ylim(9e7, 3.5e8)
plt.xlim(9e16, 3.5e23)
plt.plot(k, vg(k, 1e-20), color = "red", lw = 2)
plt.legend(["$Gruppengeschwindigkeit$"+"  ""$ vg$"], loc="upper left")
plt.xlabel("$Wellenvektor$"+"  "+"$k [m]$")
plt.ylabel("$Geschwindigkeit$" +"  "+"$[m/s]$")
plt.title("Verlauf der Gruppengeschwindigkeit")
plt.savefig("Grpgeschw.pdf")
```
