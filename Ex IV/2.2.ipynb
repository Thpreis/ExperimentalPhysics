```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
Z= np.linspace(1,100,100)

def SpectrumoverZ(Z, n):
    return np.sqrt(10973731.57)*Z*np.sqrt(1-1/n**2)
plt.plot(Z, SpectrumoverZ(Z, 2), label="n=2")
plt.plot(Z, SpectrumoverZ(Z, 3), label="n=3")
plt.plot(Z, SpectrumoverZ(Z, 4), label="n=4")
plt.title("K-Linien als Funktion der Kernladungszahl")
plt.legend()
plt.savefig("K_Linien.pdf")
plt.show()
```

```python
l = np.linspace(1e-15,1,10000000)

def Kernladungszahl(l, n):
    return (np.sqrt(10973731.57)*np.sqrt(l)*np.sqrt(1-1/n**2))**(-1)
plt.plot(l, Kernladungszahl(l, 2), label="n=2")
plt.plot(l, Kernladungszahl(l, 3), label="n=3")
plt.plot(l, Kernladungszahl(l, 4), label="n=4")
plt.title("K-Linien als Funktion der Kernladungszahl")
plt.ylim(0.,100)
plt.legend()
```
