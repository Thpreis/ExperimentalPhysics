```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

def psi1(x, s):
    return np.sqrt(1/np.sqrt(2*np.pi*s**2))*np.exp(-x**2/4/s**2)
def psi2(x, s):
    return np.sqrt(s/np.pi)*1/np.sqrt(x**2+s**2)
x = np.linspace(-20., 20., 100)
plt.plot(x, psi1(x, 2), linewidth = 2)
plt.plot(x, psi2(x, 2), linewidth = 2)
plt.ylim(0., 0.5)
plt.ylabel('Wahrscheinlichkeitsdichte')
plt.legend(["Psi 1 sigma=2", "Psi 2 sigma=2"])
plt.savefig('4_2_g1.pdf')
plt.title('Wahrscheinlichkeitsdichte der Wellenfunktionen')
plt.legend(loc='upper right')
```
