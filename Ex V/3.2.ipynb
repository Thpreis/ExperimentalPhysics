```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

a)
==

```python
#Lennard Jones Potential for two argon atoms
c=10.4 #meV
s=3.4 #Angstrom
def phi(r):
    return 4*c*((s/r)**(12)-(s/r)**(6))

r= np.linspace(3,5,1000)

plt.plot(r, phi(r), lw=1.6)
plt.ylim(-20,40)
plt.title("Wechselwirkung zweier Argonatome")
plt.ylabel("Lennard-Jones-Potential $\phi (r)$ [meV]")
plt.xlabel("Abstand r [$\AA $]")
plt.savefig("WechselwirkungsPotential.pdf")
```

```python
#Kraft zwischen zwei Argonatomen
def F(r):
    return 24*c*(2 *s**(12) /r**(13) - s**6 /r**7 )

r= np.linspace(3,7,1000)

plt.plot(r, F(r), lw=1.6)
plt.ylim(-20,40)
plt.xlim(3.4,5.7)
plt.title("Kraft zwischen zwei Argonatomen")
plt.ylabel("Kraft F(r) [meV]")
plt.xlabel("Abstand r [$\AA $]")
plt.savefig("Kraft.pdf")
```
