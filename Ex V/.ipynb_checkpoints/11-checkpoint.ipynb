```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#11.1
BC0=80.34#mT
TC=7.19#K
u=1.25e-6 #Tesla^2 m^3/J
def wärme_s(T):
    return T/u *(6*BC0**2 *T**2 *1/TC**4)
def wärme_n(T):
    return T/u*(- 2*BC0**2 *1/TC**2)
T=np.linspace(0,100,1000)
plt.plot(T,wärme_s(T), label="supra")
plt.plot(T,wärme_n(T), label="normal")
plt.legend()
    
```
