```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def Anzahldichte_Photonen(f, T):
    return f**2/(np.exp(h*f/k/T)-1)
#Gegeben:
f = np.linspace(9e12,5e15, 100)
k = 1.38e-23
h = 1.626e-34
plt.plot(f, Anzahldichte_Photonen(f, 5800), linewidth = 2, color = "red")
plt.xlabel("Frequenz f [Hz]")
plt.ylabel('Anzahldichte der Photonen')
plt.legend(["Anzahldichte bei T = 5800K"])
plt.savefig('Anzahldichte_Photonen_new.png')
```
